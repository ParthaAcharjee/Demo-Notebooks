# Linear Regression in TensorFlow
In this notebook, we'll estimate a model using linear regression in tensorflow. First, we'll generate input dataset from a model and add small noise. Then we'll try to estimate model using linear regression.

## Generate dataset

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

N=100
x=2*np.pi*np.linspace(0,0.25,N)
ymodel=np.sin(x)
y=ymodel+np.random.normal(0,0.1,N)

def plot_data_model(title):
    plt.scatter(x,y,alpha=0.2,color='darkorange',label='data')
    plt.plot(x,ymodel,color='k',label='model')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title, fontweight='bold',fontsize=16)
    plt.legend(prop={'size':20})
    plt.show()
```

## Define tensorFlow Graph

```python
# Inputs
xnode=tf.placeholder(tf.float32)
ynode=tf.placeholder(tf.float32)

# Model Paramters
w=tf.Variable(0.0)
b=tf.Variable(0.0)

ypred=w*x+b
cost=tf.reduce_mean(tf.square(ypred-y))

optimizer=tf.train.GradientDescentOptimizer(0.5).minimize(cost)
```

## Run tensorFlow Graph

```python
sess=tf.Session()
sess.run(tf.global_variables_initializer())
for i in range(50):
    if i%5==0: print("Iteration: {0:4} Cost: {1:1.4}".format(i,sess.run(cost)))
    sess.run(optimizer,feed_dict={xnode:x, ynode:y})

prediction=sess.run(ypred,feed_dict={xnode:x})
plt.plot(x,prediction,label='Prediction')
plot_data_model('Data, Model, and Prediction')
```
