# Convolution neural network in tensorboard
In this notebook, we'll implement a CNN to recognize MNIST dataset. We also display the CNN graph and parameters in tensorboard. First, we'll define few constant for this example.

## Load dataset

```python
import tensorflow as tf

from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("./MNIST_data/", one_hot=True)

imgSize=28 # Image height and width
inputs=imgSize*imgSize # Flatten array
outputs=10 # Number of classes, ten digits
channels=1 #Gray_Scale
logpath="./logs/graphCNN" # path to save tensorgraph
```

The input and output placeholders are defined inside the dataset name scope.

```python
with tf.name_scope('Dataset'):
    x=tf.placeholder(tf.float32,[None,inputs],name='Images')
    y=tf.placeholder(tf.float32,[None,outputs],name='Levels')
    
    #Reshape input image 1D array to 4D array for convulation input
    xImage=tf.reshape(x,[-1,imgSize,imgSize,channels],name='Resized')
```

## Define fucntions to create different types of layers
Now, we are going to write a function to create a convolution layer. Each convolution layer taken input from its previous layer. Description of the arguments of the function is as follows,
* **x** : input from previous layer
* **input_count**: number of inputs to the layer, number outputs of the previous layer
* **output_count**: number of filters, number of output from the concolution layer
* **filter_width**: width of the filter w. Here, width and height are same.
* **strides**: number of pixels moves at each strides. Here, vertical and horizontal strides are same.
* **layer_name**: layers are defined inside a separate name scope. It'll make our tensorgraph clean and readable.

We also add two histogram summary to get summary of weights and biases for all layersin tensorboard. 

```python
def conv_layer(x,input_count,output_count,filter_width,strides=1,layer_name='Conv'):
    with tf.name_scope(layer_name): # Put each conv layer inside a name_scope
        shape=[filter_width,filter_width,input_count,output_count]
        w=tf.Variable(tf.truncated_normal(shape,stddev=0.05),name='Weights')
        b=tf.Variable(tf.constant(0.05,shape=[output_count]),name='Biases')
        
        filter_strides=[1,strides,strides,1]
        layer=tf.nn.conv2d(x,w,filter_strides,padding='SAME')+b

        tf.summary.histogram("Weights", w)
        tf.summary.histogram("Biases", b)
        
        # We can save computation by performing maxpooling before the ReLU
        return tf.nn.relu(layer)
```

A function for the pooling layer is defined as follows, 

```python
def max_pooling(x,k=2,strides=2,layer_name='Pooling'):
    with tf.name_scope(layer_name):
        ksize=[1,k,k,1]
        stride=[1,strides,strides,1]
        layer=tf.nn.max_pool(x,ksize,stride,padding='SAME')
        
        return layer
```

After few convolution layers and few pooling layers, we have to flatten the output to feed into a fully-connected nural network. The function for flattening is defined as follows,

```python
def flatten(x, layer_name='Flat2Array'):
    with tf.name_scope(layer_name):
        tensorshape=x.get_shape()#[None, height, width, channel]
        elements=tensorshape[1:4].num_elements() 
        # flat to a 1D array
        return tf.reshape(x,[-1,elements]),elements
```

Finally, we define a method to creat a fully-connected layer.

```python
def fc_layer(x,inputs,outputs,layer_name='fc',relu=True):
    with tf.name_scope(layer_name):
        w=tf.Variable(tf.truncated_normal([inputs,outputs],stddev=0.05),name='Weights')
        b=tf.Variable(tf.constant(0.05,shape=[outputs]),name='Biases')
        
        tf.summary.histogram("Weights", w)
        tf.summary.histogram("Biases", b)

        layer=tf.matmul(x,w)+b
        if relu: return tf.nn.relu(layer)
        return layer
```

## Define CNN structure
Now, we use our previously defined functions to create the CNN. In this example, we have two convolution layers, and each of those are followed by a pooling layer. Then, we flatten the output and feed to a fully-connected neural network system. The fully-connected neural netwrk system have two layers. So, the basic structure of the CNN is follows,

**images-->>Conv_one-->>Pool_one-->>Conv_two-->>Pool_two-->>Flatten-->>FC_one-->FC_two-->>labels**

```python
with tf.name_scope('Convolution_neural_network'):
    
    conv1_outputs=16; conv1_w_width=14
    conv2_outputs=36; conv2_w_width=7
    fc1_outputs=128
                        
    Conv1=conv_layer(xImage,channels,conv1_outputs,conv1_w_width,layer_name='Conv1')
    Pool1=max_pooling(Conv1,layer_name='Pool1')
    
    Conv2=conv_layer(Pool1,conv1_outputs,conv2_outputs,conv2_w_width,layer_name='Conv2')
    Pool2=max_pooling(Conv2,layer_name='Pool2')
    
    flat_input,fc1_inputs=flatten(Pool2,layer_name='Flatten')
    
    fc1=fc_layer(flat_input,fc1_inputs,fc1_outputs,layer_name='fc1')
    fc2=fc_layer(fc1,fc1_outputs,outputs,layer_name='fc2',relu=False)
```

We also create two other blocks to optimize the cost and measure the performance.

```python
with tf.name_scope('Optimizer'):
    cross_entropy=tf.nn.softmax_cross_entropy_with_logits(logits=fc2,labels=y)
    cost=tf.reduce_mean(cross_entropy)
    optimizer = tf.train.AdamOptimizer(learning_rate=1e-4).minimize(cost)
    
with tf.name_scope('Performance'):
    y_pred=tf.nn.softmax(fc2)
    correct_pred=tf.equal(tf.argmax(y_pred,1),tf.argmax(y,1))
    accuracy=tf.reduce_mean(tf.cast(correct_pred,tf.float32))

tf.summary.scalar("Cost_summary", cost)
tf.summary.scalar("Accuracy_summary", accuracy)
```

## Train the system
We start the tensorflow session and initialize all variables. We create a summary writer object and merge all summary in a single object. We optimize the system using the training dataset in batch mode. Also, the summaries are added to the tensorboard after each fifty iterations.

```python
init = tf.global_variables_initializer()
sess=tf.Session()
sess.run(init)

train_writer = tf.summary.FileWriter(logpath, sess.graph) # for 1.0
allSummary = tf.summary.merge_all()


batch_size=100
maxIteration=1000
for itr in range(maxIteration):
    x_batch, y_batch = mnist.train.next_batch(batch_size)
    batch_dictionary={x:x_batch,y:y_batch}
    sess.run(optimizer,feed_dict=batch_dictionary)
    if itr%50==0: 
        acc,summary=sess.run([accuracy,allSummary],feed_dict=batch_dictionary)
        print('Iteration: {0} Training Accuracy: {1:.2%}'.format(itr+1,acc))
        train_writer.add_summary(summary,itr+1)
```

Finally, we measure the performance on test dataset in batch mood. Each batch contains 100 test images.

```python
test_accuracy=[]     
for batch in range(100):
    x_batch, y_batch = mnist.test.next_batch(batch_size)
    batch_dictionary={x:x_batch,y:y_batch}
    acc=sess.run(accuracy,feed_dict=batch_dictionary)
    test_accuracy.append(acc)
    if batch%10==0: print('Test Batch: {0:3} Accuracy: {1:.2%}'.format(batch+1,acc))

print('\nAverage test accuracy: {0:.2%}'.format(sum(test_accuracy)/len(test_accuracy)) )
print("\nFinished.\ntensorboard --logdir="+logpath)
```

```python
import matplotlib.pyplot as plt
plt.plot(test_accuracy)
plt.xlabel('Test batch')
plt.ylabel('Accuracy')
plt.show()
```
